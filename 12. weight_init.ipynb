```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
# 가중치 초깃값(0.01, Xavier 초깃값, He 초깃값 비교
x = np.random.randn(1000,100)  # 1000개의 데이터
node_num = 100   # 각 은닉층의 노드(뉴런) 수
hidden_layer_size = 5   # 은닉층이 5개
activations = {}   # 이곳에 활성화 결과(활성화값)를 저장
```

```python
for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]
        
    # w = np.random.randn(node_num, node_num) * 1  # vanishing gradient
    # w = np.random.randn(node_num, node_num) * 0.01  # 표현력 문제
    w = np.random.randn(node_num, node_num) / np.sqrt(node_num)  # xavier 초깃값
    a = np.dot(x, w)
    z = sigmoid(a)
    activations[i] = z
```

```python
# 층이 5개가 있으며, 각 층의 뉴런은 100개씩이다.
# 입력 데이터로서 1000개의 데이터를 정규분포로 무작위 생성하여 5층 신경망에 흘린다.
# 활성화 함수로는 시그모이드 함수를 이용
# 각 층의 활성화 결과를 activations 변수에 저장
# 가중치의 분포에 주의하자
# 이번에는 표준편차가 1인 정규분포를 이용
# 이 분포된 정도를 계속 바꿔가며 활성화값들의 분포가 어떻게 변화하는지 관찰하자
```

```python
# vanishing gradient 문제발생
for i, a in activations.items():
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "_layer")
    plt.hist(a.flatten(), 30, range=(0,1))
plt.show()
```

```python
# vanishing gradient 문제는 발생하지 않지만 시그모이드 함수의 표현력의 문제가 발생
for i, a in activations.items():
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "_layer")
    plt.hist(a.flatten(), 30, range=(0,1))
plt.show()
```

```python
# xavier 초기값 설정
for i, a in activations.items():
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "_layer")
    plt.hist(a.flatten(), 30, range=(0,1))
plt.show()
```

```python
# 활성화 함수로 ReLU를 사용할 때는 He 초깃값을, sigmoid나 tanh 등의 S자 모양 곡선일때는
# Xavier 초깃값쓰는것이 현재의 모범 사례이다.
```
